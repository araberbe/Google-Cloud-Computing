## 4. Cloud Storage: Almacenamiento de Datos No Estructurados 游늭

游늷 **쯈u칠 es Cloud Storage?**  
**Cloud Storage** es un servicio escalable para almacenar datos no estructurados, como PDFs, im치genes o videos, ideal para respaldos, distribuci칩n de contenido y archivos contables.

### Almacenamiento por Objetos

| Clase | Frecuencia de Acceso | Costo | Ejemplo Contable |
